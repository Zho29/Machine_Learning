```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import gridspec
```

```python
data = pd.read_csv("creditcard.csv")
print(data.head())
```

```python
print(data.describe())
```

```python
#Fraud and Valid Cases
fraud = data[data['Class'] == 1 ]
valid = data[data['Class'] == 0 ]

outlierFraction = len(fraud)/float(len(valid))
print(outlierFraction)
```

```python
print("Fraud Cases:",len(data[data['Class'] == 1 ]))
print("Valid Cases:",len(data[data['Class'] == 0 ]))
```

```python
print("Amount details of the fraudulent transaction:")
fraud.Amount.describe()
```

```python
print("Amount details of the valid transaction:")
valid.Amount.describe()
```

```python
#Look for the correlation betn features and target
corr=data.corr()
fig=plt.figure(figsize=(12,9))
sns.heatmap(corr,vmax=.8,square=True)
plt.show()
```

```python
#Assign for column for the target and feature 
x = data.drop(['Class'],axis=1)
y = data["Class"]
print(x.shape)
print(y.shape)
```

```python
#Prepare the data for the feature value and target value 
x=x.values
y=y.values
```

```python
#Preparing Data 

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Build and Train the Model 

from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)

y_pred=rfc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,matthews_corrcoef,confusion_matrix

accuracy=accuracy_score(y_test,y_pred)
precision=precision_score(y_test,y_pred)
recall=recall_score(y_test,y_pred)
f1=f1_score(y_test,y_pred)
mcc=matthews_corrcoef(y_test,y_pred)

print("Model Evaluation Matrics: ")
print("Accuracy: ",accuracy)
print("Precision: ", precision)
print("Recall: ", recall)
print("F1_Score",f1)
print("Mathews Correlation Coefficient: ",mcc)

```

```python
conf_matrix = confusion_matrix(y_test,y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(conf_matrix,annot=True, fmt = "d" , cmap="Blues",xticklabels=["Normal","Fraud"],yticklabels=["Normal","Fraud"])
plt.title("Confusion matrix")
plt.xlabel("Predicated Class")
plt.ylabel("True Class")
plt.show()
```
