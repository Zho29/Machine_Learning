```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python
dataset=datasets.load_diabetes(as_frame=True,scaled=False)
df=dataset.frame
print(df.head())
```

```python
df.info()
```

```python
input_df=dataset.data
input_df.info()

input_feature=dataset.feature_names
print(input_feature)
```

```python
disease_progression=dataset.target
print(disease_progression)
```

```python
corr_matrix=df.corr()
plt.figure()
plt.figure(figsize=(12, 10))
sns.heatmap(
    corr_matrix, 
    annot=True,     
    cmap='coolwarm', 
    fmt='.2f'        
)
plt.title('Correlation Matrix of Diabetes Data', fontsize=16)
plt.show()
```
